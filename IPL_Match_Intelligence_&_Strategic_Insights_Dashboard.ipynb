```python
# Calculate Team Efficiency Table

team_stats = pd.DataFrame({
    'Matches Played': matches_played,
    'Matches Won': matches_won
})

team_stats['Win %'] = (
    team_stats['Matches Won'] /
    team_stats['Matches Played'] * 100
).round(2)

team_stats.sort_values(
    'Win %',
    ascending=False
)
```
